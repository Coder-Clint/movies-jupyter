```python
movies.Film = movies.Film.astype('category')
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
```

```python
movies.info()
```

```python
gen_cat = movies.Genre.cat.categories
```

```python
list1 = []
mylabl = []
for gen in gen_cat:
  list1.append(movies[movies.Genre ==gen].BudgetMillions)
  mylabl.append(gen)
h = plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabl)
plt.legend()
plt.show()
```
